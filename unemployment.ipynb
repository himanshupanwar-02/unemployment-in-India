```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heat_maps = df_clean[['Estimated Unemployment Rate','Estimated Employment', 'Estimated Labour Participation Rate']]
heat_maps = heat_maps.corr()

#plotting the figure
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True);
```

### insight of this chat
Unemployment and Employment have a weak negative correlation (-0.22) — as unemployment rises, employment slightly falls.


Unemployment and Labour Participation have almost zero correlation (0.0026) — they do not move together.


Employment and Labour Participation also show no meaningful correlation (0.011) — changes in one don’t affect the other.


Overall, the three variables are mostly independent — no strong linear relationships exist in this dataset.

The vertical black lines (error bars) represent standard deviation or confidence intervals.

3. Seaborn barplot() is showing the mean by default

seaborn.barplot() automatically computes and displays the mean of the selected variable (Estimated Unemployment Rate) for each category (Region).

```python
plt.figure(figsize=(10,5))
plt.scatter(df_clean['Estimated Labour Participation Rate'],
            df_clean['Estimated Unemployment Rate'])

plt.title("Labour Participation vs Unemployment")
plt.xlabel("Labour Participation Rate (%)")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

### insights of this chart
No strong linear relationship
The points are widely scattered with no clear upward or downward trend. This means labour participation rate does not have a strong direct connection to unemployment rate.

High unemployment occurs even at mid participation levels
Many points between 40%–60% participation still show unemployment rates as high as 50–80%, indicating that higher participation does not guarantee lower unemployment.

Most data is clustered at lower unemployment
A dense cluster of points is seen around:

Labour Participation: 40%–60%

Unemployment Rate: 0%–20% This means most regions have moderate participation and relatively low unemployment.

Outliers with extremely high unemployment
There are some clear outlier points where unemployment shoots above 60–80%, regardless of participation rate. These indicate specific regions or periods with extreme unemployment conditions.

```python
states = df_clean['states'].unique()[:5]  # limit to 5 states for visibility

plt.figure(figsize=(12,6))
for s in states:
    data = df_clean[df_clean['states'] == s]
    plt.plot(data['Date'], data['Estimated Unemployment Rate'], label=s)

plt.title("Unemployment Trend for First 5 States")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

### insight of this chat
Bihar shows the highest unemployment spike Among the five states, Bihar has a sharp peak around 50–60%, which is significantly higher than the others. This indicates a period of severe unemployment volatility in Bihar.

Delhi also shows noticeable fluctuations Delhi’s line shows a mid-level spike around 40–45%. This suggests Delhi experienced temporary unemployment surges, though not as extreme as Bihar.

Andhra Pradesh and Assam remain relatively stable These two states show lower and smoother unemployment trends, without extreme peaks. Their unemployment rates appear more controlled and predictable.

Chhattisgarh shows a rising trend Chhattisgarh’s line gradually increases over time, indicating a steady upward trend rather than sudden spikes. This might point to long-term unemployment challenges developing in the state
